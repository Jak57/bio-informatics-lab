# 1 A - Compute the Number of Times a Pattern Appears in a Text

```python
def pattern_count(text, pattern):
    n = len(text)
    k = len(pattern)
    cnt = 0
    for i in range(n-k+1):
        kmer = text[i:i+k]
        if kmer == pattern:
            cnt += 1
    return cnt

if __name__ == "__main__":
    # text = "GCGCG"
    # pattern = "GCG"
    # print(pattern_count(text, pattern))
    with open('rosalind_ba1a.txt') as file:
        f = file.read().split("\n")
        text = f[0]
        pattern = f[1]
        print(pattern_count(text, pattern))
```

# 1 B - Find the Most Frequent Words in a String

```python
def most_frequent_kmers(text, k):
    n = len(text)
    dic = {}
    for i in range(n-k+1):
        kmer = text[i:i+k]
        if kmer in dic.keys():
            dic[kmer] += 1
        else:
            dic[kmer] = 1

    mx = max(dic.values())
    most_freq = []
    for kmer in dic:
        if dic[kmer] == mx:
            most_freq.append(kmer)
    return most_freq

if __name__ == "__main__":
    # text = "ACGTTGCATGTCGCATGATGCATGAGAGCT"
    # k = 4
    # print(most_frequent_kmers(text, k))
    with open('rosalind_ba1b.txt') as file:
        f = file.read().split("\n")
        text = f[0]
        k = int(f[1])
        ans = most_frequent_kmers(text, k)
        for kmer in ans:
            print(kmer, end=" ")
```

# 1 C - Find the Reverse Complement of a String

```python
def reverse_complement(dna):
    dna = dna[::-1]
    dic = {
        'A': 'T',
        'C': 'G',
        'G': 'C',
        'T': 'A'
    }
    rev_c = ""
    for ch in dna:
        rev_c += dic[ch]
    return rev_c

if __name__ == "__main__":
    # dna = "AAAACCCGGT"
    # print(reverse_complement(dna))
    with open('rosalind_ba1c.txt') as file:
        f = file.read().split("\n")
        dna = f[0]
        print(reverse_complement(dna))
```

# 1 D - Find All Occurrences of a Pattern in a String

```python
def all_occurences(pattern, dna):
    pos = []
    n = len(dna)
    k = len(pattern)
    for i in range(n-k+1):
        sub = dna[i:i+k]
        if sub == pattern:
            pos.append(i)
    return pos

if __name__ == "__main__":
    # pattern = "ATAT"
    # dna = "GATATATGCATATACTT"

    with open('rosalind_ba1d.txt') as file:
        f = file.read().split("\n")
        pattern = f[0]
        dna = f[1]
    
    pos = all_occurences(pattern, dna)
    for idx in pos:
        print(idx, end=" ")
```

# 1 E - Find Patterns Forming Clumps in a String

```python
def find_clump(dna, k, L, t):
    n = len(dna)
    clumps = set()

    for i in range(n-L+1):
        window = dna[i:i+L]
        dic = {}
        for j in range(L-k+1):
            kmer = window[j:j+k]
            if kmer not in dic.keys():
                dic[kmer] = 1
            else:
                dic[kmer] += 1

        for kmer in dic:
            if dic[kmer] >= t:
                clumps.add(kmer)
    return clumps

if __name__ == "__main__":
    # dna = "CGGACTCGACAGATGTGAAGAAATGTGAAGACTGAGTGAAGAGAAGAGGAAACACGACACGACATTGCGACATAATGTACGAATGTAATGTGCCTATGGC"
    # k, L, t = 5, 75, 4

    with open('rosalind_ba1e.txt') as file:
        f = file.read().split("\n")
        dna = f[0]
        tmp = f[1].split(" ")
        k = int(tmp[0])
        L = int(tmp[1])
        t = int(tmp[2])

    clumps = find_clump(dna, k, L, t)
    for clump in clumps:
        print(clump, end=" ")
```

# 1 F - Find a Position in a Genome Minimizing the Skew

```python
def skew(genome):
    diff = []
    g, c = 0, 0
    for ch in genome:
        if ch == 'G':
            g += 1
        elif ch == 'C':
            c += 1
        diff.append(g-c)
    
    mn = min(diff)
    pos = []
    for i, val in enumerate(diff):
        if val == mn:
            pos.append(i+1)
    return pos

if __name__ == "__main__":
    #genome = "CCTATCGGTGGATTAGCATGTCCCTGTACGTTTCGCCGCGAACTAGTTCACACGGCTTGATGGCAAATGGTTTTTCCGGCGACCGTAATCGTCCACCGAG"
    
    with open('rosalind_ba1f.txt') as file:
        f = file.read().split("\n")
        genome = f[0]
    
    pos = skew(genome)
    for p in pos:
        print(p, end=" ")
```

# 1 H - Find All Approximate Occurrences of a Pattern in a String

```python
def hamming_distance(dna1, dna2):
    mismatch = 0
    for i in range(min(len(dna1), len(dna2))):
        if not(dna1[i] == dna2[i]):
            mismatch += 1
    return mismatch

def approximate_pattern(pattern, text, d):
    k = len(pattern)
    n = len(text)

    dic = {}
    for i in range(n-k+1):
        kmer = text[i:i+k]
        if kmer not in dic:
            dic[kmer] = [i]
        else:
            dic[kmer].append(i)

    pos = []
    for kmer in dic:
        if (hamming_distance(pattern, kmer) <= d):
            pos = pos + dic[kmer]

    return sorted(pos)

if __name__ == "__main__":
    # pattern = "ATTCTGGA"
    # text = "CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAATGCCTAGCGGCTTGTGGTTTCTCCTACGCTCC"
    # d = 3

    with open('rosalind_ba1h.txt') as file:
        f = file.read().split("\n")
        pattern = f[0]
        text = f[1]
        d = int(f[2])

    pos = approximate_pattern(pattern, text, d)
    for p in pos:
        print(p, end=" ")
```

# 1 L - Implement PatternToNumber

```python
def pattern_to_number(pattern):
    n = len(pattern)
    dic = {
        'A': 0,
        'C': 1,
        'G': 2,
        'T': 3
    }

    pattern = pattern[::-1]
    prod = 1
    ans = 0
    for ch in pattern:
        ans += (dic[ch] * prod)
        prod *= 4

    return ans

if __name__ == "__main__":
    # dna = "AGT"

    with open('rosalind_ba1l.txt') as file:
        f = file.read().split("\n")
        dna = f[0]

    number = pattern_to_number(dna)
    print(number)
```

# 1 M - Implement NumberToPattern

```python
def number_to_pattern(num, k):
    dic = {
        '0': 'A',
        '1': 'C',
        '2': 'G',
        '3': 'T'
    }
    
    pattern = ""
    while (num > 0):
        rem = num % 4
        num = num//4
        pattern += dic[str(rem)]
    
    while not(len(pattern) == k):
        pattern += 'A'
    
    return pattern[::-1]

if __name__ == "__main__":
    # num = 45
    # k = 4

    with open('rosalind_ba1m.txt') as file:
        f = file.read().split("\n")
        num = int(f[0])
        k = int(f[1])

    print(number_to_pattern(num, k))
```

# 1 N - Generate the d-Neighborhood of a String

```python
def hamming_distance(dna1, dna2):
    mismatch = 0
    for i in range(min(len(dna1), len(dna2))):
        if not(dna1[i] == dna2[i]):
            mismatch += 1
    return mismatch

def neighborhood(kmers, pattern, d):
    nei = []
    for kmer in kmers:
        if hamming_distance(kmer, pattern) <= d:
            nei.append(kmer)
    return nei


def create(s, kmers, k):
    if len(s) == k:
        kmers.append(s)
        return s
    create(s+'A', kmers, k)
    create(s+'C', kmers, k)
    create(s+'G', kmers, k)
    create(s+'T', kmers, k)

if __name__ == "__main__":
    # pattern = "ACG"
    # d = 1

    with open('rosalind_ba1n.txt') as file:
        f = file.read().split("\n")
        pattern = f[0]
        d = int(f[1])

    kmers = []
    create('', kmers, len(pattern))
    ans = neighborhood(kmers, pattern, d)
    for kmer in ans:
        print(kmer)
            
```
